```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
data = np.genfromtxt("wmap_P_1loop.dat")
data13 = np.genfromtxt("wmap_P13.dat")
data22 = np.genfromtxt("wmap_P22.dat")
datapk = np.genfromtxt("dmeff_pk.dat")
kk = data[:, 0]
Pk = data[:, 1]
kk13 = data13[:, 0]
Pk13 = data13[:, 1]
kk22 = data22[:, 0]
Pk22 = data22[:, 1]
kk_dmeff = datapk[:,0]
Pk_dmeff = datapk[:,1]
```

```python
plt.figure()
plt.xscale('log');plt.yscale('linear'); #plt.ylim(-10**5,10**5)
#plt.xlabel(r'$\ell$')
#plt.ylabel(r'$ C_\ell^\mathrm{\Phi\Phi}$')
plt.plot(kk,Pk,'g-')
plt.show()
```

```python
plt.figure()
plt.xscale('log');plt.yscale('log'); #plt.ylim(-10**5,10**5)
#plt.xlabel(r'$\ell$')
#plt.ylabel(r'$ C_\ell^\mathrm{\Phi\Phi}$')
plt.plot(kk,Pk,'g-')
plt.plot(kk_dmeff, Pk_dmeff, 'b-')
plt.show()
```
