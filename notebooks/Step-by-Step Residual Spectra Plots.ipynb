Next, we add corrections from the 1loop_dm_reference_modified.py


```python
# get all data for EFT using 1loop_dm_reference_modified.py

# after "=" write path to the .dat file with Pk corrections

# the code I used is in PT-check/code/1loop_dm_reference_modified.py
# Output files (from .ini files above) I used are in PT-check/code/...P_1loop.dat
# for example, name1 = "/Users/aizhan.akh/Documents/Projects/PT-check/code/lcdm_P_1loop.dat"


# load corrections to Pk
name1 =                       # for LCDM corrections
name2 =                       # for IDM with n =0 corrections
name3 =                       # for IDM with n =2 corrections
name4 =                       # for IDM with n =-2 corrections
lcdm_c = np.genfromtxt(name1)
dmeff0_c = np.genfromtxt(name2)
dmeff2_c = np.genfromtxt(name3)
dmeff_2_c = np.genfromtxt(name4)
kk = lcdm_c[:, 0]
Pk = lcdm_c[:, 1]
kk1 = dmeff0_c[:, 0]
Pk1 = dmeff0_c[:, 1]
kk2 = dmeff2_c[:, 0]
Pk2 = dmeff2_c[:, 1]
kk3 = dmeff_2_c[:,0]
Pk3 = dmeff_2_c[:,1]
```

```python
# now get spectra with EFT CORRECTIONS

# we want to add Pk to lcdm values which correspond to particular kk values
# so we need to interpolte lcdm values for those particular values of kk
lcdm_interp = np.interp(kk, kk_lcdm, Pk_lcdm)
lcdm_new = Pk + lcdm_interp

# same interpolation with dmeff sets
dmeff0_interp = np.interp(kk1, kk_dmeff_0, Pk_dmeff_0)
dmeff0_new = Pk1 + dmeff0_interp

dmeff2_interp = np.interp(kk2, kk_dmeff_2, Pk_dmeff_2)
dmeff2_new = Pk2 + dmeff2_interp

dmeff_2_interp = np.interp(kk3, kk_dmeff__2, Pk_dmeff__2)
dmeff_2_new = Pk3 + dmeff_2_interp

# calculate residuals

res_new1 = (dmeff0_new-lcdm_new)/lcdm_new
res_new2 = (dmeff2_new-lcdm_new)/lcdm_new
res_new3 = (dmeff_2_new-lcdm_new)/lcdm_new
```

```python
# Compare Pk plots of linear models and models with new EFT corrections
plt.figure()
plt.xscale('log');plt.yscale('log')
plt.xlabel(r'$k \,\,\,\, [h/\mathrm{Mpc}]$')
plt.ylabel(r'$P(k) \,\,\,\, [\mathrm{Mpc}/h]^3$')
plt.plot(kk_lcdm,Pk_lcdm,'k-', label = "LCDM")
plt.plot(kk_dmeff_0,Pk_dmeff_0,'g-', label = 'n = 0')
plt.plot(kk_dmeff_2,Pk_dmeff_2,'r-', label = 'n = 2')
plt.plot(kk_dmeff__2,Pk_dmeff__2,'b-', label = 'n = -2')
plt.plot(kk, lcdm_new, 'k--',label = "LCDM, EFT")
plt.plot(kk1, dmeff0_new, 'g--',label = "n = 0, EFT")
plt.plot(kk2, dmeff2_new,'r--', label = "n = 2, EFT")
plt.plot(kk3, dmeff_2_new,'b--', label = "n = -2, EFT")

plt.legend(loc = "lower left")
plt.title("EFT CORRECTIONS VS LINEAR")
plt.show()
```

```python
#plot Residual Pk spectra of HALOFIT models and models with EFT corrections
plt.figure()
plt.xscale('log');plt.yscale('linear');plt.xlim(kk[0],kk[-1]) # the region where Pk starts to diverge noticeably - can change to [0] to [-1]
plt.xlabel(r'$k \,\,\,\, [h/\mathrm{Mpc}]$')
plt.ylabel(r'$(P(k)(\sigma_0) - P(k)(\lambda CDM))\,/\, P(k)(\lambda CDM)  \,\,\,\, [\mathrm{Mpc}/h]^3$')
plt.plot(kk_lcdm_h,res_Pk0_h,'b-', label = "n = 0, w/ halofit")
plt.plot(kk_lcdm_h,res_Pk2_h,'g-', label = "n = 2, w/ halofit")
plt.plot(kk_dmeff__2_h,res_Pk_2_h,'k-', label = "n = -2, w/ halofit")
plt.plot(kk1, res_new1, 'g-',label = "n = 0, EFT")
plt.plot(kk2, res_new2,'r--', label = "n = 2, EFT")
plt.plot(kk3, res_new3,'b--', label = "n = -2, EFT")
plt.title("Pk RESIDUAL SPECTRA")
plt.legend(loc ="lower left")
plt.show()
```
